```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Filter the data for movies released in the 1990s
# How to subset a DataFrame
movie_df = netflix_df[netflix_df["type"] == "Movie"]

# Filtering by column values
movie_df_subset = movie_df[movie_df["release_year"] > 1989]
movie_df_1990s = movie_df_subset[movie_df_subset["release_year"] < 2000]

# Find the most frequent movie duration
# Visualizing a distribution
plt.hist(movie_df_1990s["duration"])
plt.title('Distribution of Movie Durations in the 1990s')
plt.xlabel('Duration in minutes')
plt.ylabel('Number of Movies')
plt.show

duration = 100

# Count the number of short action movies from the 1990s
# Subsetting by genre
movie_df_1990s_action = movie_df_1990s[movie_df_1990s["genre"] == "Action"]

# Setting up a counter
short_movie_count = 0

# Iterating through a DataFrame
# Counting a condition
for label, row in movie_df_1990s_action.iterrows():
    if row["duration"] < 90:
        short_movie_count += 1
    
print(short_movie_count)
```
